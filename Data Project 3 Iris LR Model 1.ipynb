```python
#Point 1 import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
print('Libraries Imported')

#Point 2 Retrieve Table from csv file
data = pd.read_csv(r"E:\Cognixia\Iris.csv")
data.drop(columns=['Id'], inplace= True)
iris = data.rename({'SepalLengthCm': 'sepal_length', 
                 'SepalWidthCm': 'sepal_width', 
                 'PetalLengthCm': 'petal_length', 
                 'PetalWidthCm': 'petal_width',
                 'Species': 'species'}, axis=1)

#Point 3 Show first five records 
iris.head()

#Point 4 Show dataframe which have sepalwidth greater than 4
pd.set_option('display.max_rows', None)
iris1=iris[iris['sepal_width'] > 4]
iris1.head()

# Point 5 Show dataframe which have petalwidth greater than 1
pd.set_option('display.max_rows', None)
iris2=iris[iris['petal_width'] > 1]
iris2.head(93)

#Point 6 Reterive records which have petalwidth more than 2
pd.set_option('display.max_rows', None)
iris3=iris[iris['petal_width'] > 2]
iris3.head(23)

#Point 7 and 8 Create colored scatterplot
fig = iris[iris.species=='Iris-setosa'].plot(kind='scatter',x='sepal_length',y='petal_length',color='orange', label='Setosa')
iris[iris.species=='Iris-versicolor'].plot(kind='scatter',x='sepal_length',y='petal_length',color='blue', label='versicolor',ax=fig)
iris[iris.species=='Iris-virginica'].plot(kind='scatter',x='sepal_length',y='petal_length',color='green', label='virginica', ax=fig)
fig.set_xlabel("Sepal Length")
fig.set_ylabel("Petal Length")
fig.set_title("Sepal Length VS Petal Length")
fig=plt.gcf()
fig.set_size_inches(15,15)
plt.show()
plt.savefig('Sepal Length VS Petal Length.png')

# Model 1 point a and b create variables x and y
iris.drop(columns=['species'], inplace= True)
Y= iris.drop(labels= ['sepal_width','petal_length','petal_width'], axis= 1)
X= iris.drop(labels= ['sepal_length','petal_length','petal_width'], axis= 1)

# Model 1 point c and divide x and y variables using train test split 
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size= 0.30)
print(X_train[:5])
print(X_test[:5])
print(Y_train[:5])
print(Y_test[:5])

# Model 1 point e Create an object named as lr and assign memory from linearregression()  method.
lr = LinearRegression()

# Model 1 point f Fit both training set into fit method 
lr.fit(X_train, Y_train)

#Model 1 point g and h Predict x_test from predict method and store the result into y_pred obect and Show first five records from actual and predicted objects
lr.predict(X_test)
Y_pred = lr.predict(X_test).sum(axis=1)
print(Y_pred[:5])

#Model 1 point i Evaluating Model 1's Performance lower mse
print('Mean Absolute Error:', mean_absolute_error(Y_test, Y_pred))
print('Mean Squared Error:', mean_squared_error(Y_test, Y_pred))
print('Mean Root Squared Error:', np.sqrt(mean_squared_error(Y_test, Y_pred)))
```
